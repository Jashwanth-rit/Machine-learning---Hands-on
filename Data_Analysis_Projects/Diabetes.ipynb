```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file
data = pd.read_csv('../diabetes.csv')
```

```python
# Display the first few rows
print(data.head())
```

```python
# Step 1: Cleaning the Data
# Replacing zeros with NaN for specific columns where 0 is likely invalid
columns_to_clean = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
data[columns_to_clean] = data[columns_to_clean].replace(0, np.nan)
```

```python
# Imputing missing values with the median of each column
data.fillna(data.median(), inplace=True)
```

```python
# Step 2: Summary Statistics
print("Dataset Summary:")
print(data.describe())
```

```python
# Step 3: Visualization
# Histograms for all features
data.hist(bins=15, figsize=(15, 10), color='#6A5ACD', edgecolor='black')
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Outcome Analysis: Pairplots
sns.pairplot(data, hue="Outcome", diag_kind="kde", corner=True)
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

```python
# Boxplot to visualize outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=data[columns_to_clean])
plt.title("Boxplot for Continuous Features")
plt.show()
```
