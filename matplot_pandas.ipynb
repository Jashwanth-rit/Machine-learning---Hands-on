```python
import pandas as pd
import numpy as np
from numpy.random import randn

%matplotlib inline

#create some data 

d  = pd.DataFrame(randn(50,4),columns=['sun','mon','tue','wed'])
```

```python
d
```

```python
d['mon'].hist(bins = 10)
```

```python
d['sun'].plot(kind = 'hist',bins = 20)
```

```python
d['mon'].plot(kind = 'hist')
```

```python
d['sun'].hist(grid = False)
```

```python
#Area plot 
#Here abs() is used because of negative values present in table that shows an error
#kind is area , hist , many other option we can give,
#alpha used to make color lighter
#legend , title are already known to u
#we can get info regarding paramter that can be passed through plot by 
#placing cursor inside the bracket and press shift+tab
#To get only specified column just pass the column name in d ( that is pd array)



d.abs().plot(kind = 'area',alpha = 0.5,legend = True,title = "this is my plot")
```

```python
#Bar graph - 

d.abs().plot(kind = 'bar')
```

```python
#with stacked

d.abs().plot(kind = 'bar',stacked = True,)
```

```python
#Other way to do that is

d.abs().plot.bar(stacked=True)
```

```python
#Line - charts
#lw - line width
#figsize - graph size
#we can get info regarding paramter that can be passed through plot by 
#placing cursor inside the bracket and press shift+tab


d.abs().plot(kind = 'line',stacked = True,lw = 2,figsize = (10,5))
```

```python
#Scatter plots
#x,y is to be specified else shows error , and x,y are the column name 
#u can use matplotlib website for colors if multiple columns are added 
#Due to minimum number of points graph looks wierd
#cmap is used to give color for points
#c - means color

d.abs().plot(kind = 'scatter',stacked = True,x = 'mon',y = 'tue',c = 'wed',cmap = 'Greys')


```

```python
#instead of c use s - size

d.abs().plot(kind = 'scatter',x = 'mon',y = 'tue',s = d['wed']*200,alpha  = 0.3)
```

```python
#Box graph

d.abs().plot(kind = 'box',cmap = 'jet')
```

```python
#Hex - plot
#gridsize increases size of hexgon in graph - by increasing or zoom in the graph

d.abs().plot(kind = 'hexbin',x = 'mon',y = 'tue',gridsize = 7 ,cmap ='Greys' )
```

```python
#Kde ploting

d.plot(kind = 'kde')
```

```python
#density plot

d.plot(kind = 'density')
```
