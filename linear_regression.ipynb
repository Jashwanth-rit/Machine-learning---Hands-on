```python
#Linear regression

import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes

# Load our diabetes dataset
diabetes  =  load_diabetes()

# Convert the data to a pandas DataFrame
data = pd.DataFrame(diabetes.data,columns = diabetes.feature_names)

# Add the target 
data['target'] = diabetes.target

# Grab the head
data.head(5)

```

```python
# Split the data in x and y

# Add the features - mean age sex all this to x .
# Add the target column on which u need to predict to y .

x = data.drop('target',axis = 1)
y = data['target']

#print the shape of data that are splited

#x,y
print('the data in x is :\n ',x)
print('the data in y is :\n ',y)
print('the shape of x is : \n',x.shape)
print('the shape of y is :\n ',y.shape)
```

```python
# traning and testing phase -  train x and y by 80% of data
                          # -  test x and y by 20% of data

from sklearn.model_selection import train_test_split

# Split the data into training and testing

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
# just initialing test_size is enough because traing will occupy remaining space or data

# Print the shape of the training data and testing data of x, y

print('the data in x_train is : \n',x_train)
print('the data in x_test is : \n',x_test)
print('the data in y_train is :\n ',y_train)
print('the data in y_test is : \n',y_test)


print('the shape of x_train is :\n ',x_train.shape)
print('the shape of x_test is :\n ',x_test.shape)
print('the shape of y_train is : \n',y_train.shape)
print('the shape of y_test is :\n ',y_test.shape)
```

```python
# Linear aggression for data 

from sklearn.linear_model import LinearRegression

#Create a Linear Regression Model
lr = LinearRegression()


# Train the model on the training set -  this is how the model learns
lr.fit(x_train, y_train)

# Predict on the testing set 
# Y_prods are the predicted values of the target variable based on the features data
y_pred = lr.predict(x_test)
# Once we get the prediction , we can compare them to the actual value , 
#by this we get performance of model

y_pred
```

```python
# To know whether the model is getting good or not , we need to make graph 

# Making some matrics
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Evalute the performance of the model

# variance of the dependent variable(target) explained by features data
# How well the model fits the data , higher is the better
r2 = r2_score(y_test,y_pred)

#MSE : Average distance between the predicted and the actual values. lower is better
mse = mean_squared_error(y_test,y_pred)

#MAE : Average absolute distance between the predicted and actual values , lower is better
mae = mean_absolute_error(y_test,y_pred)

# Intercept : storing points of the regression line on the y-axis.
# the value of dependent variable (target ) when the independent variables
# If positive number , target increases as features increase , if negative nuber , target decreases as feature incrses
intercept = lr.intercept_

```

```python
import matplotlib.pyplot as plt

#plot the predicted vs actual values
# predicted values of the target vaiables are plotted on x - axis
# Actual values of target are ploted on y - axiz
# if the model is a good fit , the pints should be close to the diagonal line,
# indicatesing , a strong linear relationship between  the actual and predicted value

plt.scatter(y_test,y_pred,alpha = 0.5)
plt.plot([y.min(),y.max()],[y.min(),y.max()], color = 'red')
plt.xlabel("Actual values")
plt.ylabel("predicted values")
plt.title("predicted vs actual values (checking the performance of the model)")
plt.show()
```

```python
# Plot the residuals 
# the residuals are plotted against the predicted values of the target variable
# if the linear regression model is good fit for the data,
# the residual plot should show a random scatter of the points around zero
# with no discernable trend.

plt.scatter(y_pred,y_test - y_pred,alpha = 0.5)

plt.xlabel("predicted values")
plt.ylabel("residuals")
plt.title("Residulas plot (checking the performance of the model)")
plt.show()
```

```python
# Core of the classifier 

# Biulding our own classifier
from scipy.spatial import distance

# Assume that we have plotted the graph 
# a , b are the two points on the graph , which we have plotted using the training data
# This returns the distance between the a and b
def euc(a,b):
    return distance.euclidean(a,b);

# Classifier class
class ScappyKNN():

    # fit is used to assign the values to variables
    def fit(self,train_features,train_labels):
        self.features = train_features
        self.labels = train_labels

    # Predict will return the predicted value for all the test_features
    # Predicted value mean nearest value -
    # we have assumed that we plotted the graph in that graph
    # we find the nearest value to all individual test_features
    
    def predict(self, test_features):
        prediction = []
        for row in test_features:
            label = self.closet(row)
            prediction.append(label)
        return prediction

    # THis function returns the nearest points distance for the passed row
    # check The nearest distance between row and all the points and returns the minimum at distance 
    def closet(self,row):
        nearest_distance = euc(row,self.features[0])
        nearest_idx = 0
        for i  in range(1,len(self.features)):
            dist = euc(row,self.features[i])
            if dist < nearest_distance :
                nearest_distance = dist
                nearest_idx = i
        return self.labels[nearest_idx]
        



# Machine learning with Iri s data set

# Step 1 : importing data or handling data

# Import Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
# iris.feature_names # this names features_names and target_names can be available in documentation of sklearn
# iris.target_names
# iris .data[0] # features in first row
# iris.target[0] # labels in first row 
# pd.DataFrame(iris)

# Step 2 : splting data into training and testing 

test_idx = [0,50,100] # all this index are the first index of the three diffrent flowers

# Trainign data
train_target = np.delete(iris.target,test_idx)  # To delete testing Data from iris table and add targets (labels) to training target
train_data = np.delete(iris.data,test_idx,axis = 0)   # To delete testing Data from iris table and add data (features) to training data

# Testing data
test_target = iris.target[test_idx] # adding the testing data to test split ( the data which is delted from adding to train is used for testing to check the accuracy of the model)
test_data = iris.data[test_idx]

# Step 3 : trainig of the model
clf = ScappyKNN()
clf.fit(train_data,train_target)

# Step 4 : testing the model by using testing data passign through predict
prediction = clf.predict(test_data)  # on passing the test_data it should give output of test_target , then its performance is perfect

# To find accuracy or check the performance
from sklearn.metrics import accuracy_score
accuracy_score(test_target,prediction)
```
