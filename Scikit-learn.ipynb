```python
from sklearn.datasets import fetch_california_housing

# Fetch the California housing dataset
housing = fetch_california_housing()

# X (features) and y (target)
X = housing.data  # This contains the features
y = housing.target  # This contains the target prices
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
```

```python
mod = KNeighborsRegressor()
mod.fit(X,y)
```

```python
pred = mod.predict(X)
# Printing actual target values and predictions
print("Actual target values (y):\n", y, "\n")
print("Predicted values (pred):\n", pred, "\n")
```

```python
import matplotlib.pylab as plt
```

```python
import numpy as np

# Scatter plot for actual vs predicted values
plt.scatter(y, pred, label='Predicted vs Actual', alpha=0.5)

# Plot the ideal linear line (y = x) to represent perfect prediction
# This assumes that the range of y and pred is similar
line_range = np.linspace(min(y), max(y), 100)
plt.plot(line_range, line_range, color='red', label='Ideal line (y = x)', linestyle='--')

# Adding labels and title
plt.xlabel('Actual Values (y)')
plt.ylabel('Predicted Values (pred)')
plt.title('Actual vs Predicted Values with Linear Line')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('scale',StandardScaler()),
    ('model',KNeighborsRegressor()),
])
```

```python
pipe.fit(X, y)
```

```python
pred = pipe.predict(X)
```
