```python
#To remove the row which consist of Nan 
my_df.dropna()
```

```python
#To remove the row or column consist of minimum threshold amount of Nan 
#For-row
my_df.dropna(thresh=2)
#For-column
my_df.dropna(thresh=2,axis=1)

#Note: for column we should specify axis , for row we need not to specify axis
```

```python
#To fill specified value in place of Nan 
# we use a function called fillna

my_df.fillna(value = 10)

#since there is no Nan value present ,table not going to change
```

```python
#To set mean value of that column in place of Nan

my_df.fillna(value = my_df['id'].mean())

#To set min value of that column to Nan

my_df.fillna(value = my_df['id'].min())

#To set min value of that column to Nan

my_df.fillna(value = my_df['id'].max())
```
